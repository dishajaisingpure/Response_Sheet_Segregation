```python
# Importing all the necessary packages

import pandas as pd
import glob
import re
```

```python
# Getting the input file from the folder and reading its data

for files in glob.iglob(r".\Response Sheet processing\*.xlsx"):    #Specify the location of the files
    data = pd.read_excel(files) 
#The column names required for sorting are extracted using regex and stored in a variable 
ss = re.findall('........ivision',str(data))[0]       # The column name is Select Division-(Case insensitive)
ss1 = re.findall('...........roll....',str(data))[0]  # The column name is Enter your roll no.-(Case insensitive)
print(ss,ss1)
data = data.sort_values([ss])  #Sorting data division-wise
data['Score'] = data['Score'].map(str) + ' '+ '/ 15'  # Adding the total score to the score column
print(data)
```

```python
# Writing the data in the output file using the xlsxwriter and then setting the column width for all rows in the workbook

writer = pd.ExcelWriter(r".\Response Sheet processing\response sheet output.xlsx", engine='xlsxwriter')
for i in data[ss].unique():
    df = data.loc[data[ss] == i]
    df = df.sort_values([ss1]) #Sorting data roll no.-wise
    print(df)
    df.to_excel(writer, sheet_name = i, index=False)
for worksheet in writer.sheets.values():
    worksheet.set_column(0,worksheet.dim_colmax,25)
writer.save()
```
